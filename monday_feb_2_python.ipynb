```python
num1 = float(input("Enter first number:  "))
num2 = float(input("Enter second number:  "))
op = input("Choose numerical operation (e.gg., +, -, *, etc.): ")

if op == "+":
    res = num1 + num2
    print(f"Result: {res}")
elif op == "-":
    res = num1 - num2
    print(f"Result: {res}")
elif op == "*": # Corrected from '+'
    res = num1 * num2
    print(f"Result: {res}")
elif op == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero.")
    else:
        res = num1 / num2
        print(f"Result: {res}")
elif op == "//":
    if num2 == 0:
        print("Error: Cannot perform floor division by zero.")
    else:
        res = num1 // num2
        print(f"Result: {res}")
elif op == "**":
    res = num1 ** num2
    print(f"Result: {res}")
elif op == "%":
    if num2 == 0:
        print("Error: Cannot perform modulo by zero.")
    else:
        res = num1 % num2
        print(f"Result: {res}")
else:
    print("Invalid operation")
```
