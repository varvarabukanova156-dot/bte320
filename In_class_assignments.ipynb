1: Income Statement

```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = sales * 0.40
gross_profit = sales - cogs
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

print("Sales: $", format(sales, ",.2f"))
print("Cost of Goods Sold: $", format(cogs, ",.2f"))
print("Gross Profit: $", format(gross_profit, ",.2f"))
print("Taxes: $", format(taxes, ",.2f"))
print("Net Income: $", format(net_income, ",.2f"))
```

Movie Theatre

```python
import math

print("Ad$ Profit")

for ad in range(0, 201, 25):
    attendees = 20 + 2 * math.sqrt(ad)
    revenue = attendees * 10
    profit = revenue - ad - 200
    print(float(ad), round(profit, 1))
```

Rock Paper Scissors

```python
import random

play = input("Do you want to play? yes/no: ")

while play == "yes":
    user = input("Enter rock, paper, or scissors: ")
    computer = random.choice(["rock", "paper", "scissors"])

    print("Computer chose:", computer)

    if user == computer:
        print("Tie")
    elif user == "rock" and computer == "scissors":
        print("You win")
    elif user == "paper" and computer == "rock":
        print("You win")
    elif user == "scissors" and computer == "paper":
        print("You win")
    else:
        print("Computer wins")

    play = input("Do you want to play again? yes/no: ")
```

Recursive String Reversal

```python
def reverse(s):
    if len(s) == 0:
        return s
    else:
        return reverse(s[1:]) + s[0]


print(reverse("hello"))
print(reverse("12345"))
```

ROI Calculator

```python
def roi(cost, revenue):
    if type(cost) not in [int, float] or type(revenue) not in [int, float]:
        return "Inputs must be numeric"

    if cost <= 0:
        return "Cost must be greater than 0"

    result = (revenue - cost) / cost
    return result


print(roi(100, 120))
```

Loan Payment Calculator

```python
def loan_payment(principal, annual_rate, years):
    if principal <= 0 or annual_rate <= 0 or years <= 0:
        return "All inputs must be positive"

    r = annual_rate / 12
    n = years * 12

    payment = (r * principal) / (1 - ((1 + r) ** (-n)))
    return payment


print(loan_payment(10000, 0.06, 5))
```

Quality Control Manager (exceptions)

```python
def getValues(n):
    alist = []

    for i in range(n):
        while True:
            try:
                value = float(input("Enter a value: "))

                if value < 0:
                    print("Invalid input")
                else:
                    alist.append(value)
                    break
            except:
                print("Invalid input")

    return alist


def calcAverage(alist):
    new_list = alist[3:]
    avg = sum(new_list) / len(new_list)
    return avg


def main():
    n = int(input("How many values? "))
    alist = getValues(n)
    avg = calcAverage(alist)
    print("Average excluding first three values =", avg)


main()
```

Greeting Class

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal):
        if formal == True:
            print(self.last + ", " + self.first)
        else:
            print(self.first)

    def __str__(self):
        return self.first + " " + self.last


g = Greeting("Varya", "Bukanova")
g.display(True)
g.display(False)
print(g)
```

Student Class

```python
class Student:
    def __init__(self, name, number, courses):
        self.name = name
        self.number = number
        self.courses = courses

    def getName(self):
        return self.name

    def getCourses(self):
        return self.courses

    def setName(self, new_name):
        self.name = new_name

    def __len__(self):
        return len(self.courses)

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)


s = Student("Varya", 101, ["BTE320", "MAS201"])
print(s.getName())
print(s.getCourses())
s.setName("Varya B")
print(s.getName())
print(len(s))
s.enroll("FIN302")
print(s.getCourses())
```
